```python
import numpy as np
import matplotlib.pyplot as plt
depth = np.loadtxt('input_1hii_dummy.ovr')[:,0] #for T= 30000 K
HII=np.loadtxt('input_1hii_dummy.ovr')[:,7]

plt.plot(depth, HII) 
  
# naming the x axis 
plt.xlabel('depth') 
# naming the y axis 
plt.ylabel('ion-fraction') 
```
